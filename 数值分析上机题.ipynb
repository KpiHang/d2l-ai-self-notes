## 插值法

### 绘制Lagrange插值函数图像

已知函数在3,1,4的值分别是4，6，9，试构造Lagrange插值多项式

3个点构造2次插值多项式；

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 展示数据点
x = np.array([3,1,4])
y = np.array([4,6,9])
plt.scatter(x, y)
```

```python
def lagrange(x, y, num_points, x_test):
    # 所有的基函数值，每个元素代表一个基函数的值
    l = np.zeros(shape=(num_points, ))
    # 计算第k个基函数的值
    for k in range(num_points):
        # 乘法时必须先有一个值
        # 由于l[k]肯定会被至少乘n次，所以可以取1
        l[k] = 1
        # 计算第k个基函数中第k_个项（每一项：分子除以分母）
        for k_ in range(num_points):
            # 有些资料上显示k是不等于k_的
            if k != k_:
                # 基函数需要通过连乘得到
                l[k] = l[k]*(x_test-x[k_])/(x[k]-x[k_])
            else:
                pass 
    # 计算当前需要预测的x_test对应的y_test值        
    L = 0
    for i in range(num_points):
        # 求所有基函数值的和
        L += y[i]*l[i]
    return L
```

```python
x_test = list(np.linspace(1, 4, 50))
y_predict = [lagrange(x, y, len(x), x_i) for x_i in x_test]
```

```python
plt.scatter(x, y)
plt.plot(x_test, y_predict, c="r", label="Lagrange Func")
plt.legend()
```
