自己造的带噪音的数据，来体验过拟合，并使用正则化进行解决；

```python
%matplotlib inline
import torch
import numpy as np
from torch import nn
from d2l import torch as d2l
```

$$y = 0.05 + \sum_{i = 1}^d 0.01 x_i + \epsilon \text{ where }
\epsilon \sim \mathcal{N}(0, 0.01^2).$$

模拟样本y，自己给加一点噪音（N(0, 0.01^2)）

```python
# 训练集容量，测试集容量， 特征维数（几个特征）  
n_train, n_text, num_inputs, batch_size = 20, 100, 200, 5 # 强模型，小数据量，容易过拟合；
# 一个特征对应一个权重； 这是无噪音的w ， b
true_w, true_b = torch.ones((num_inputs, 1)) * 0.01, 0.05
train_data =d2l.synthetic_data(true_w, true_b, n_train) # d2l中特有的根据w，b形状，合成出n_train大小的样本
train_iter = d2l.load_array(train_data, batch_size) # 按批量从训练集中拿出一个batch_size的样本；
test_data = d2l.synthetic_data(true_w, true_b, n_text)
test_iter = d2l.load_array(test_data, batch_size, is_train=False) # 测试集不用来训练模型；
```

```python
for X_, y_ in train_iter:
    print(X_, y_)
    break
# X_ 5*200, y_ 5*1   
```

## 从零开始实现

- 初始化模型参数

```python
def ini_params():
    w = torch.normal(0, 1, size=(num_inputs, 1), requires_grad=True)
    b = torch.zeros(1, requires_grad=True)
    return [w, b]
```

- 定义L2范数惩罚项


```python
def l2_penalty(w):
    return torch.sum(w.pow(2)) / 2 

# lambd 放在train()中了
```

- 定义训练代码实现

```python
# lambd 在训练过程中是个固定值
def train(lambd):
    w, b = ini_params()
    net, loss = lambda X : d2l.linreg(X, w, b), d2l.squared_loss # , 之前是lambda匿名函数
    num_epochs, lr = 100, 0.03
    # 动画效果展示结果
    animator = d2l.Animator(xlabel='epochs', ylabel='loss', yscale='log',
                            xlim=[5, num_epochs], legend=['train', 'test'])
    for epoch in range(num_epochs):
        for X, y in train_iter: # X 和 y是一个batch_size的大小的样本，不是单个样本
            l = loss(net(X), y) + lambd * l2_penalty(w)
            l.sum().backward() # 所以这里要sum一下； backward() 自动计算梯度了；
            d2l.sgd([w, b], lr, batch_size) # sgd中params.grad 直接调用grad
        if(epoch + 1) % 5 == 0:
            animator.add(epoch + 1, (d2l.evaluate_loss(net, train_iter, loss),
                                     d2l.evaluate_loss(net, test_iter, loss)))
    print('w的L2范数是：', np.linalg.norm(w.detach().numpy())) 
    # np.linalg.norm()用于求范数 默认ord=2 L2范数；
```

- lambd = 0 忽略正则化直接训练

```python
train(lambd=0)
```

- lambd = 3 的正则化训练模型

```python
train(lambd=3)
```

## 简洁实现

```python
# 解析请看笔记；
def train_concise(wd):
    net = nn.Sequential(nn.Linear(num_inputs, 1))
    
    for param in net.parameters():
        param.data.normal_()
        
    loss = nn.MSELoss(reduction="none")
    
    num_epochs, lr = 100, 0.005
    
    trainer = torch.optim.SGD([
        {"params":net[0].weight,'weight_decay': wd},
        {"params":net[0].bias}], lr=lr)
    
    animator = d2l.Animator(xlabel='epochs', ylabel='loss', yscale='log',
                            xlim=[5, num_epochs], legend=['train', 'test'])
    
    for epoch in range(num_epochs):
        for X, y in train_iter:
            trainer.zero_grad()
            l = loss(net(X), y)
            l.mean().backward()
            trainer.step() # 更新参数
        if (epoch + 1) % 5 == 0:
            animator.add(epoch + 1,
                         (d2l.evaluate_loss(net, train_iter, loss),
                          d2l.evaluate_loss(net, test_iter, loss)))
    print('w的L2范数：', net[0].weight.norm().item())
```

- lambd = 0 不使用正则化

```python
train_concise(0)
```

- lambd = 3 的正则化

```python
train_concise(3)
```
